```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)

np.random.seed(sum(map(ord, "categorical")))
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```python
#非常实用的方法是将Seaborn的分类图分为三类，将分类变量每个级别的每个观察结果显示出来，
#显示每个观察分布的抽象表示，以及应用统计估计显示的权重趋势和置信区间：
#第一个包括函数swarmplot()和stripplot()
#第二个包括函数boxplot()和violinplot()
#第三个包括函数barplot()和pointplt()
```

```python
# 分类散点图
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True); #jitter表示加入随机抖动，使得点不在同一条直线上
```

```python
# 使用函数swarmplot()，它使用避免重叠点的算法将分类轴上的每个散点图点定位，使用jitter还是可能有重叠：
sns.swarmplot(x="day", y="total_bill", data=tips);
```

```python
# 也可以传入hue参数添加多个嵌套的分类变量。高于分类轴上的颜色和位置时冗余的，现在每个都提供有两个变量之一的信息：
sns.swarmplot(x="day", y="total_bill", hue="sex",data=tips);
```

```python
# 当x轴变量名比较长时，可以反转x,y轴
sns.swarmplot(y="day", x="total_bill", hue="sex",data=tips);
```

```python
#在类别之间进行简单比较并汇总信息可是使用一下函数
```

```python
# 箱线图
#sns.set(rc={"figure.figsize": (10, 8)})
sns.boxplot(x="day", y="total_bill", hue="time", data=tips);
```

```python
# 提琴图
sns.violinplot(y="total_bill", x="day", hue="time", data=tips);
```

```python
# 类别内的统计估计
# 条形图
sns.barplot(x="sex", y="survived", hue="class", data=titanic);
```

```python
#当您想要显示每个类别中的观察次数，而不是计算第二个变量的统计量。
sns.countplot(x="deck", data=titanic, palette="Greens_d");
```

```python
# 反转
sns.countplot(y="deck", hue="class", data=titanic, palette="Greens_d");
```

```python
# 点图
# pointplot()函数提供了可视化相同信息的另一种风格。该函数还对另一轴的高度估计值进行编码，
# 而不是显示一个完整的柱型，它只绘制点估计和置信区间
sns.pointplot(x="sex", y="survived", hue="class", data=titanic);
```

```python
# 绘制“宽格式”数据
sns.boxplot(data=iris,orient="h");
```

```python
# 绘制多层面板分类图
#可以使用更高级别的函数factorplot()，将以上这些函数与FacetGrid()相结合，默认情况下，factorplot()产生一个pairplot()
sns.factorplot(x="day", y="total_bill", hue="smoker", data=tips);
```

```python
sns.factorplot(x="day", y="total_bill", hue="smoker", data=tips, kind="bar");
```

```python
# 调用"facet"展开更多其他分类变量
sns.factorplot(x="day", y="total_bill", hue="smoker",
               col="time", data=tips, kind="swarm");
```

```python
# 基于FacetGrid的工作原理，要更改图形的大小和形状，需要指定适用于每个方面的size和aspect参数：

sns.factorplot(x="time", y="total_bill", hue="smoker",
               col="day", data=tips, kind="box", size=4, aspect=.5);
```
