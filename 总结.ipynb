```python
df_tips = sns.load_dataset("tips")
```

# distplot( )  kdeplot( )
## 直方图、密度曲线图


```python
df_iris = sns.load_dataset("iris")
fig, axes = plt.subplots(1,2)
sns.distplot(df_iris['sepal_length'], ax = axes[0], kde = True, rug = True)        # kde 密度曲线  rug 边际毛毯  
sns.kdeplot(df_iris['sepal_length'], ax = axes[1], shade=True)                     # shade  阴影                         
```

```python
sns.set( palette="muted", color_codes=True)  
rs = np.random.RandomState(10)  
d = rs.normal(size=100)  
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)  
sns.distplot(d, kde=False, color="b", ax=axes[0, 0])  
sns.distplot(d, hist=False, rug=True, color="r", ax=axes[0, 1])  
sns.distplot(d, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])  
sns.distplot(d, color="m", ax=axes[1, 1])  
```

# pointplot画出变量间的关系

```python
grouped_df = df_tips.groupby('size')['total_bill'].aggregate(np.median).reset_index()
plt.figure(figsize=(12,8))
sns.pointplot(grouped_df['size'].values, grouped_df.total_bill.values, alpha=0.8)
plt.ylabel('Median total_bill', fontsize=12)
plt.xlabel('size number', fontsize=12)
plt.xticks(rotation='vertical')
```

# barplot()

```python
f, ax=plt.subplots(figsize=(6,6))

#orient='h'表示是水平展示的，alpha表示颜色的深浅程度
sns.barplot(y=grouped_df['size'].values, x=grouped_df.total_bill.values,orient='h', alpha=0.8, color='red')

#设置y轴、X轴的坐标名字与字体大小
plt.ylabel('size', fontsize=16)
plt.xlabel('total_bill', fontsize=16)

#设置X轴的各列下标字体是水平的
plt.xticks(rotation='horizontal')

#设置Y轴下标的字体大小
plt.yticks(fontsize=16)
```

# bar图

```python
plt.rc('font', family='SimHei', size=13)

num = np.array([13325, 9403, 9227, 8651])
ratio = np.array([0.75, 0.76, 0.72, 0.75])
men = num * ratio
women = num * (1-ratio)
x = ['聊天','支付','团购\n优惠券','在线视频']

width = 0.5
idx = np.arange(len(x))
plt.bar(idx, men, width, color='red', label='男性用户')
plt.bar(idx, women, width, bottom=men, color='yellow', label='女性用户')  #这一块可是设置bottom,top，如果是水平放置的，可以设置right或者left。
plt.xlabel('应用类别')
plt.ylabel('男女分布')
plt.xticks(idx+width/2, x, rotation=40)

#bar图上显示数字

for a,b in zip(idx,men):

    plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=12)
for a,b,c in zip(idx,women,men):
    plt.text(a, b+c+0.5, '%.0f' % b, ha='center', va= 'bottom',fontsize=12)

plt.legend()
```

```python
#读取的时候,将索引列变为chn_name,这样画图时候X轴自动为索引
df=df_tips
df.index.name='size'#这样x轴的label就变成‘size了’。

plt.rc('font', family='SimHei', size=13)
plt.figure()
df['total_bill'].plot(kind='bar')
plt.ylabel('total_bill')
plt.title('total_bill')

p = df['tip']
p.plot(color='black',secondary_y=True,style='--o',linewidth=2)#style--表示虚线，-表示实线
plt.ylabel('tip')
x=[0,1,2,3,4]#因为x轴是汉字，所以默认对应的数值是从0开始的
for a,b in zip(x,p):
    plt.text(a+0.1, b+0.02, '%.2f' % b, ha='center', va= 'bottom',fontsize=12)
```

```python
education=df['size'].value_counts()
df_education=pd.DataFrame({'education':education.index[1:],'fre':education.values[1:]})
df_education.index=df_education.education
plt.figure()
df_education.fre.plot(kind='bar')  
plt.ylabel('计数')
plt.xlabel('size') 
plt.title('size分布')
```

# 画饼状图

```python
#根据value_counts（）结果画饼图
phone=df.phone_operator.value_counts()
df_phone=pd.DataFrame({'phone_operator':phone.index[1:],'fre':phone.values[1:]})

plt.rc('font', family='SimHei', size=13)
fig = plt.figure() 
plt.pie(df_phone.fre,labels=df_phone.phone_operator,autopct='%1.2f%%') #画饼图（数据，数据对应的标签，百分数保留两位小数点） 
plt.title("手机运营商分布") 
```
