```python
%matplotlib inline 
#jupyter notebook 中的魔法函数，如果不是使用该软件请使用plt.show()用于显示图像
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord,"aesthetics")))  # 定义种子
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,10000)
    for i in range(1,7):
        plt.plot(x,i*x)
```

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
```

```python
import seaborn as sns # matplotlib的扩展包
sinplot()
```

```python
# seaborn 将matplotlib参数分成两个独立的组，第一组设定美学风格，第二组设定不同的度量元素
# 第一组：操作这些参数的接口是两对函数。为了控制样式，使用axes_style()和set_style()函数
#第二组：通过 plotting_context() 和 set_context() 调整绘图元素
```

```python
#5个seaborn的主题 
'''
    darkgrid 黑色网格（默认）
    whitegrid 白色网格
    dark 黑色背景
    white 白色背景
    ticks 应该是四周都有刻度线的白背景？
'''
sns.set_style("darkgrid")
sinplot()
```

```python
'''
特别的可以通过sns.axes_style(style=None, rc=None) 返回一个sns.set_style()可传的参数的字典
通过类似sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})的方式写入更具体的配置样式。
'''
```

```python
### 用despine()进行边框控制
#despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(left=True,offset=10, trim=True); #left:删除左边框;offset两坐标轴离开距离;trim参数会限制留存的边框范围
```

```python
# 使用with语句中套用axes_style()达到临时设置参数的效果。这也允许创建不同风格的坐标轴。
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
# 查看set_style()包含哪些参数
sns.axes_style()
```

```python
# 通过 plotting_context() 和 set_context() 调整绘图元素
# 通过sns.set()重置参数
sns.set()
```

```python
# 四种预设，按相对尺寸的顺序(线条越来越粗)，分别是paper，notebook(默认), talk, and poster。
sns.set_context("poster")
plt.figure(figsize=(10,6))
sinplot()
```

```python
# 通过更改context还可以独立地扩展字体元素的大小。(这个选项也可以通过顶级set()函数获得）。
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1.5})
sinplot()
```
